## <span style="color: #9C68C8;">Text mit Python</span> 
### <span style="color: #9C68C8;">Text abspeichern</span> 

Das Speichern von Textdateien (Dateiendung `.txt`) kann man sich wie das Ablegen von Sachen in einem Fach eines Regals vorstellen. In dem Fall ist das Regal das Speichersystem des Computers, und das bestimmte Fach des Regals ist die Datei, an der wir arbeiten. Nach der Logik machen wir also 3 Schritte zum Abspeichern von Text:
1. Fach öffnen: `open()`
2. Daten ablegen: `write()`
3. Fach wieder schließen: `close()`

In der Praxis sieht das dann so aus:

````python
datei = open("text.txt", "a")
datei.write("Lies mich!")
datei.close()
````

Jetzt taucht unsere kleine Datei `text.txt` links in der Dateiübersicht auf. Auch hier kannst du draufklicken, um sie dir anzuschauen. Schreibe doch noch etwas dazu!

🔎 <span style="color: #2E88B6;"> Unter der Haube macht der Befehl `plt.savefig()` auch nichts anderes, als die Datei anzulegen, die Datein hineinzuschreiben und die Datei dann wieder zu "schließen". Das passiert allerdings alles automatisiert in diesem einen Befehl, sodass wir uns nicht darum kümmern müssen.</span>

### <span style="color: #9C68C8;">Text einlesen</span> 

````python
datei = open("text.txt", "r")
inhalt = datei.read()
print(inhalt)
datei.close()
````

Wenn du im Schritt davor noch etwas in die Textdatei dazugeschrieben hast, dann kannst du jetzt sehen, dass es hier auftaucht.

🔎 <span style="color: #2E88B6;"> Wenn du dir den Code oben anguckst, sieht er fast so aus wie davor. Beim Befehl `open()` benutzen wir jetzt aber den Buchstaben `r` statt `a`, wie beim Abspeichern einer Datei. Das ist auch gut so: Das `a` steht für "append" (englisch für "anhängen" oder "dazuschreiben" in unserem Fall) und das `r` steht für "read" (englisch für "lesen", "einlesen").</span>

**<span style="color: #EB6ECB;">Aufgabe 7.3:</span>** *Die Daten unten kennen wir schon. Wir haben sie in Woche 6 verwendet, um uns die Indizierung von NumPy-Arrays näherzubringen. Verwende sie nun, um sie in eine Textdatei zu schreiben. Schreibe sie dabei im Format:*

`Wochentag Temperatur`

🔎 <span style="color: #2E88B6;">Tipps: Verwende eine for-Schleife. Eine neue Zeile kann mit dem string `"\n"` geschrieben werden. Strings lassen sich einfach mit `+` aneinanderhängen.</span>

````python
wochentage = ["Montag", "Dienstag", "Mittwoch", "Donnerstag", "Freitag", "Samstag", "Sonntag"]
temperaturen_wochentage = ["25.5", "26.8", "24.3", "27.1", "23.9", "17.5", "18.5"]

datei = open("daten.txt", "a")
for i in range(len(wochentage)):
    datei.write(wochentage[i] + " " + temperaturen_wochentage[i] + "\n")
datei.close()
````

**<span style="color: #EB6ECB;">Aufgabe 7.4:</span>** *Schreibe das kleine Einmaleins in eine Textdatei.*

🔎 <span style="color: #2E88B6;">Tipp: Mit `str()` kann man aus einer Zahl einen String machen, den man dann in die Datei schreiben kann.</span>

````python
datei = open("einmaleins.txt", "a")
for i in range(1,10):
    datei.write(str(i) + " * " + str(i) + " = " + str(i*i) + "\n")
datei.close()
````

## <span style="color: #9C68C8;">Tabellen mit Python</span> 

Wir bleiben bei dem Beispiel mit den Wochentagen und erweitern die Daten nur um eine weitere Woche. Also so:


```python
temperaturen_wochentage = np.array([[25.5, 26.8, 24.3, 27.1, 23.9, 17.5, 18.5], [20.3, 19.3, 22.1, 23.5, 25.5, 24.9, 21.9]])
```

In welchem Format könnten sich die Daten noch speichern lassen? Wir haben bereits die Bilddatei kennengelernt sowie gerade die Textdatei. In der Praxis würde man die Daten aber wahrscheinlich in einer Tabelle speichern. Das hat zwei Vorteile:

1. bessere Übersichtlichkeit: Es ist eindeutig erkennbar, welchem Wochentag welche Temperatur zugeordnet ist. So, wie es in dem Code dargestellt ist, ist das nicht sofort ersichtlich.
2. bessere Verarbeitbarkeit: Tabllenformate sind stark standardisiert und funktionieren mit vielen unterschiedlichen Programmen. Deshalb braucht man sich zum Beispiel keine Sorgen zu machen, die Daten irgendwann mal von Hand abtippen zu müssen, weil sie nicht einlesbar sind.

Oft werden Tabellen as sogenannte *csv*-Dateien abgespeichert, also als Dateien mit der Endung `.csv`. Diese Dateien sind tatsächlich einfach nur Textdateien, bei denen die Werte in einer Zeile durch ein Komma getrennt werden. Daher kommt auch der Name: *comma separated values - englisch für Komma-getrennte Werte*. Das ganze sieht dann ungefähr so aus:

<center>

Nun wollen wir also die untenstehenden Daten als Tabelle abspeichern. Das Werkzeug dafür haben wir schon kennengelernt: NumPy eignet sich hervorragend dafür. Dessen Grundformate, die *NumPy-Arrays*, haben wir ja auch schon kennengelernt und sind also wieder mal unser Ausgangspunkt für diese Aufgabe:

````python
temperaturen_wochentage = np.array([[25.5, 26.8, 24.3, 27.1, 23.9, 17.5, 18.5], [20.3, 19.3, 22.1, 23.5, 25.5, 24.9, 21.9]])
````

Die Funktion, die wir dafür verwenden, heißt `.savetxt()`. Es gibt noch einige andere Methoden, die Tabellen zu speichern, die wären für uns aber erstmal aufwendiger als `.savetxt`.

Wir übergeben der Funktion dann einfach die Informationen, die sie braucht:
1. Dateiname, in dem das Ganze abgespeichert werden soll: `"temperaturen.csv"`
2. Unser NumPy-Array: `temperaturen_wochentage`
3. Das Format, in dem die Zahlen geschrieben werden sollen: Wir wollen eine Gleitkommazahl (*float*) mit einer Nachkommastelle abspeichern. Das bedeutet für den Computer: `"%.1f"`
4. Das Trennzeichen (*delimiter*), durch welches unsere Werte getrennt werden sollen. Typischerweise ist das ein Komma (*comma separated value*)

````python
np.savetxt("temperaturen.csv", temperaturen_wochentage, fmt="%.1f", delimiter=",")
````

Nach Ausführen der oberen Zelle solltet ihr die Datei wieder links im Dateibrowser sehen.

Nun machen wir dasselbe wie vorher mit der Textdatei auch: Öffne die Datei und ändere einen der Werte. Dann lesen wir die veränderte Datei wieder ein:

````python
temperaturen_wochentage_veraendert = np.loadtxt("temperaturen.csv", delimiter=",")

temperaturen_wochentage_veraendert
    
````

**<span style="color: #EB6ECB;">Aufgabe Z.1:</span>** *Unten stehen die Messdaten einer Gruppe von 11 Versuchspersonen, deren Körßergröße und -gewicht dokumentiert wurden. Erstelle ein Streudiagramm, in dem du die Größe gegen das Gwicht aufträgst. Speichere das Bild als eine Datei ab. Lässt sich ein Zusammenhang erkennen?*

````python
gewicht = np.array([39.70, 45.44, 53.36, 58.08, 63.86, 68.94, 75.72, 82.64, 88.80, 92.24, 100.48])
groesse = np.array([1.33, 1.49, 1.46, 1.46, 1.73, 1.81, 1.59, 1.79, 1.79, 1.76, 1.94])


plt.scatter(gewicht, groesse)
plt.xlabel("Gewicht in kg")
plt.ylabel("Körpergröße in m")
plt.title("Körpergröße aufgetragen gegen das Gewicht")
plt.savefig("koerpergroesse_gewicht.jpg", format="jpg")
````


**<span style="color: #EB6ECB;">Aufgabe Z.2:</span>** *Schreibe nun die Daten in eine Textdatei, mit dem Format:*

`Körpergröße Gewicht`

*Schreibe in die erste Zeile noch, was wofür steht!*

````python
datei = open("versuchsdaten.txt", "a")

datei.write('Körpergröße Gewicht\n')
for i in range(len(groesse)):
    datei.write(str(groesse[i]) + " " + str(gewicht[i]) + "\n")
datei.close()
````
